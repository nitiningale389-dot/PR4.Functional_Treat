```python
print("Welcome to the Data Analyzer and Transformer Program \n")
 
print("1. Input Data ")
print("2. Display Data Summary ")
print("3. Calculate factorial ") 
print("4. Filter data by threshold ")
print("5. Sort data")
print("6. Display dataset statistics ")
print("7. Exit program ")


global_data = []

while True :
  
  

  def option_1():
    """1. input data manually (1D)"""
    print(option_1.__doc__)
    global global_data
    n = input("Enter tottal element :")
    ff = [int(x) for x in n.split()]
    print(f"Data successfully saved {global_data}\n")
    global_data.extend(ff)
 


  def option_2():
     """you selected display data summary"""
     print(option_2.__doc__)
     print("Data summary :\n")
     global global_data
     print(f"-Total elements :{len(global_data)}")
     print(f"-Maximum value : {max(global_data)}")
     print(f"-Minimum value : {min(global_data)}")
     print(f"-Sum of all value : {sum(global_data)}")
     print(f"-Avg of all  value : {sum(global_data)/len(global_data)}")


  def option_3(q):
       """you selected calculate factorial of elements (recursion)"""
      
       if q <= 1:
         return 1 
       else :
        return q * option_3(q-1)


 
  def option_4(threshold):
       """ filter data by threshold """
       print(option_4.__doc__)
       global global_data
       if not global_data:
           print("dataset is empty! nothing to filter.")
           return
       
       f_data = [x for x in global_data if x >= threshold]  
       print(f"Filtered data (values >= {threshold}): {f_data}")        
       
 


  def option_5():
       """ You selected option 5 sort data """
       print(option_5.__doc__)
       global global_data
       if not global_data:
          print("data set is empty! nothing to sort ")
          return
      
       sort = int(input("Choose sorting option :\n 1. Ascending \n 2. Descending"))

      
       if sort == 1:
          print(f"Sorted data in ascending order :{sorted(global_data)}")
       if sort == 2:
          print(f"Sorted data in descending order :{sorted(global_data,reverse=True)}")
       else:
           print("Enter choice from 1 or 2 ") 



  option_6 = lambda data:print( 
       #global global_data
      f"Dataset statistics :\n"
      f"Miximum value : {max(data) if data else "data set is empty!"}\n"
      f"Minimum value :{min(data) if data else "data set is empty!"}\n"
      f"Sum of all  values :{sum(data) if data else "data set is empty!"}\n"
      f"Average  value :{sum(data)/len(data) if data else "data set is empty!"}\n"
      #print(option_6.__doc__)
      )
  option_6.__doc__ = " Display dataset statistics "

         

              
              

          
 



       
     
     
  
  chc = int(input("Enter your choice : "))
 
  if chc == 1:
       option_1()


  elif chc == 2:
        option_2()  

  elif chc == 3: 
       q = int(input("Enter num to calculate factorial :"))
       ans = option_3(q)
       print(f"Factorial of{q} is : {ans}")
       option_3(q)   

  elif chc == 4:
       num = int(input("Enter a threshold value to filter out data above this value : "))
       option_4(num)

  elif chc == 5:
        option_5()

  elif chc == 6:
        print(option_6.__doc__)
        option_6(global_data)


  elif chc == 7:
      
      print("Exit")
      break


  else :
       print(" Select any option from 1 to 6  ")
          


 

  

     
```
